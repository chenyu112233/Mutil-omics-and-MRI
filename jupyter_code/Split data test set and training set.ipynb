```python
dir="C:\\Users\\24993\\Desktop\\jupyter\\data"

exp=pd.read_excel(os.path.join(dir,"grey60_ttest.xlsx"))
meta=pd.read_excel(os.path.join(dir,"meta.xlsx"))
MRI_features=pd.read_excel(os.path.join(dir,"MRI_features_boruata.xlsx"))
concat_data=pd.read_excel(os.path.join(dir,"Intergrate_data.xlsx"))
label=pd.read_excel(os.path.join(dir,"label.xlsx"))

exp=exp.iloc[:,1:]
meta=meta.iloc[:,1:]
MRI=MRI_features.iloc[:,1:]
concat_data=concat_data.iloc[:,1:]
y=label.iloc[:,1:]
```

```python
AUC=0
acc=0
y_test1=0
y_train1=0
for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(concat_data, y, test_size=0.3,stratify=y)
    model=LogisticRegression(max_iter=5000)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    accuracy = accuracy_score(y_test, y_pred)
    AUC1=roc_auc_score(y_test, y_pred_proba)
    if AUC1> AUC:
        AUC=AUC1
        acc=accuracy
        y_train1=y_train
        y_test1=y_test

    
#         print(roc_auc_score(y_test, y_pred_proba))
print("AUC",accuracy)
print("ROC",AUC)
```
